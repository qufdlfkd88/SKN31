##  원하는 형태(shape) 텐서 생성
- **torch.tensor(자료구조 \[, dtype\])**
    - 지정한 dtype(Data type)에 맞는 Tensor객체를 생성해서 반환한다.
      
## 특정 타입의 Tensor를 직접 생성
- torch.tensor()로 생성하면서 dtype을 지정하면 아래 타입의 Tensor객체가 생성된다.

- **tensor.shape, tensor.size(\[축번호\])**
    -  tensor의 shape조회
- **tensor.dtype, tensor.type()**

- **tensor.ndim, tensor.dim()**  : tensor 차원
- **tensor.numel()**: 전체 원소 개수

## 동일한 값을 원소로 가지는 Tensor 생성
- **torch.zeros(\*size), zeros_like(텐서)**: 0으로 구성된 tensor 생성
- **torch.ones(\*size), ones_like(텐서)**: 1로 구성된 tensor생성
- **torch.full(size, fill_value), full_like(텐서, fill_value)**: 지정한 값으로 구성된 tensor생성
  - size 
    - shape을 지정한다.
    - zeros, ones는 가변인자이므로 축별 size를 순서대로 서정
    - full 은 tuple/list로 shape을 지정한다.
    

```python
a = torch.zeros(3,2,3) # 3 x 2 x 3
a.dtype
torch.ones(2,3)  # 2 X 3
# torch.full([3,2], fill_value=100) # 3 x 2
```

## 난수를 이용한 생성

- **torch.rand(\*size)**: 0 ~ 1사이 실수로 구성된 배열을 생성. 각 값은 균등분포를 따른다.
- **torch.randn(\*size)**: 표준정규분포(평균:0, 표준편차:1)를 따르는 실수로 구성된 배열 생성
- **torch.normal(mean=평균, std=표준편차, \*size)**: 정규분포를 따르는 실수로 구성된 배열 생성
- **torch.randint(low=0, high, size)**: 지정한 범위의 정수로 구성된 배열 생성
- **torch.randperm(n)**: 0 ~ n-1 사이의 정수를 랜덤하게 섞은 값을 원소로 가지는 배열 생성

```python
torch.manual_seed(0)  # seed 설정
# torch.rand(1, 3, 5) # 1 x 3 x 5       0 ~ 1 사이 실수
# torch.randn(3, 5) # 3 X 5
# torch.normal(mean=10, std=20, size=[4, 2, 3]) # 평균 10 , 표준 편차 20
# torch.randint(1, 100, (3, 3, 6))  # 1 ~ 100 정수
torch.randperm(100) # 0 ~ 99 를 섞어서 구성
```

    - 장치는 문자열로 설정한다.
        - CPU 사용: "cpu"
        - nvida GPU: "cuda"
        - Apple m1: "mps"
            - pytorch 1.12 부터 지원
- 옮기기
    - tensor 생성시 `device` 파라미터를 이용해 설정

# Reshape

## shape 변경
- 원소의 개수가 변하는 shape으로는 reshape이 안된다.
- tensor객체.reshape(\*shape) / view(\*shape) 이용
    - 변환 후 값을 변경하면 원본 배열의 값도 같이 바뀐다.
 > tensor.clone(): tensor를 복제한다.
   - 원본 값이 변하지 않게 하려면 복사 해야한다

```python
a=torch.rand(12) # reshape 할때는 무조건 12개가 나와야 한다
a2 = a.reshape(3,4)
a3 = a.reshape((3,2,2))
a4 = a.reshape((3,2,-1))  #한 개 axis는 -1로 설정가능하고 그럼 계산해서 알아서 설정해 준다.
print(a.shape, a2.size(), a3.shape, a4.shape)
               
```

```python
a5 = a.view(3,4)
a6 = a.view((3,2,2))
a7 = a.view((3,2,-1))  #한개 axis는 -1로 설정가능
print(a.shape, a5.size(), a6.shape, a7.shape)
```

```python
# a5[0, 0] = 12.1
a2[0, 1] = 15.1
print(a2)
# print(a)
# print(a5)
```

```python
# tensor복사: clone() 메소드
r = a.reshape(3, 4).clone()
r[0,0] = 100.1
print(a)
print(r)
```

# 값의 위치(index) 변경
- tensor 원소의 index의 위치를 바꾼다.
- `tensor.transpose(axis1, axis2)` 
    - 두 축의 자리만 변경 할 수 있다.
- `tensor.permute(axis1, axis2, axis3, ..)`
    - 두 개 이상의 축 자리를 변경한다.

```python
X = torch.arange(24).reshape(2, 3, 4)
print(X.shape)

y = X.transpose(1, 2)      # 2,3,4 에서 1번과 2번의 자리를 바꿈 0번은 유지함
print(y.shape)

z = X.permute(2, 0, 1)     # 2  ,  3  ,  4     ->   4  ,  2  ,  3
print(z.shape)             # 0번   1번   2번
```

```python
# 원소별 처리
x=torch.arange(-4, 5).reshape(3,3)
print(x)
print(torch.abs(x)) # 절대값 
print(torch.sqrt(torch.abs(x))) #  제곱근
print(torch.exp(x))  # torch.e**x
print(torch.log(torch.abs(x)))
print(torch.log(torch.exp(torch.tensor(1))))  # torch.log() 밑이 e인 로그계산
print(torch.log10(torch.tensor(10)))         # torch.log10() 밑이 10인 로그계산
print(torch.log2(torch.tensor(2)))           # torch.log2() 밑이 2인 로그계산
```

```python
y = x + torch.randn((3,3))
print(y)
print(torch.round(y)) # 반올림
print(torch.round(y, decimals=2)) # 소수점 둘째자리 이하에서 반올림
print(torch.floor(y)) # 내림
print(torch.ceil(y)) # 올림
```

### 행렬곱
- <font color='red'>`@` 연산자 또는 `torch.matmul(tensor1, tensor2)` 함수 이용</font>

```python
x = torch.FloatTensor([[1, 2],[3, 4],[5, 6]])

y = torch.FloatTensor([[1, 2],[1, 2],])
print(x.size())    # x의 열수
print(y.size())    # y의 행수 가 같기에 행렬곱 가능
```

```python
z1 = x @ y
z2 = torch.matmul(x, y)
print(z1.shape, z2.shape)
print(z1)
print(z2)
```

```python
# Batch 행렬곱(Batch matrix muliplication) - bmm()
# 피연산자로 두개의 3차원 tensor를 axis (1, 2)를 기준으로 행렬곱을 처리한다.
import torch
x = torch.FloatTensor(3,4,2)    # 차원이 3으로 같아야하고 , x = (4,2), y = (2,5) x의 2열과 y의 2행이 같아야함
y = torch.FloatTensor(3,2,5)
z = torch.bmm(x, y)
z.shape
```

# autograd(자동미분)
- 자동 미분을 이용해 gradient(미분계수)를 계산하는 pytorch system.
- 딥러닝 모델에서 weight와 bias tensor들(Parameter)은 backpropagation(역전파)를 이용해 gradient를 구해서 loss가 줄어드는 방향으로 update를 하게된다.
- pytorch는 이런 미분 수행을 자동으로 처리해 준다.
    - gradient(기울기)를 구한다는 것은 미분을 한다는 것을 말한다.
- tensor가 미분 가능하려면(gradient 계산 대상 변수) `requires_grad=True` 로 설정되 있어야 한다. (default: False)

> - **미분**
>   -  (순간) 변화율을 계산한다.
>   - $$\frac{\partial y}{\partial x}$$
>       - x에 대한 y의 변화율. 즉 x가 변하면 y는 얼마나 변할까?

```python
import torch
```

```python
x = torch.tensor([1.0], requires_grad=True) # requires_grad=True 미분 계산을 해야 한다 알려 주는 것 (False는 미분 계산할 필요가 없다는 것)
print(x)

print("x의 값(데이터):",x.data)
print("계산된 X의 gradient의값:", x.grad)
print("Gradient를 계산하기 위함 도함수:", x.grad_fn)
```

```python
y = x ** 2
print(y)
# 함수의 피연산자에 requires_grad=True인 Tensor가 있으면 
# 이 함수의 도함수(gradient 계산함수)를 생성해서 결과를 저장하는 
# tensor의 grad_fn 속성에 저장한다.
```

```python
y.grad_fn(torch.tensor([2.0, 3, 7, 2]))
```

```python
# gradient를 계산
y.backward() # dy / dx
# 계산된 gradient 값을 x(requires_grad=True)의 grad 속성에 저장한다.
```

```python
x.data, x.grad
```

```python
x = torch.tensor([2.0], requires_grad=True) # 미분할 대상 알림
y = torch.tensor([3.0]) # requires_grad=False
z = x - y # 결과: z.data, x에 대한 도함수: z.grad_fn
z.backward() # x에 대한 grad값을 z.grad_fn을 이용해서 계산한 뒤 x.grad에 저장.
z
```

```python
x.data, x.grad
```

```python
y.data, y.grad
```

```python
# 선형회귀 모델 대상으로 적용
input_data = torch.tensor([30.2]) # 30.2에 0.2를 곱함 , input_data는 x값이므로 weight에  requires_grad=True 를 줌
weight = torch.tensor([0.2], requires_grad=True)

y_pred = weight * input_data
print(y_pred)
print(y_pred.grad_fn)
```

```python
print(weight.grad)
y_pred.backward()
print(weight.grad)
```

```python
# new weight
new_weight = weight.data - weight.grad # 학습률
new_weight
```

```python
# 편미분
# 하나의 함수에 gradient를 계산해야하는 변수가 두개 이상일 경우 변수별로 따로 계산한다.
# 하나의 변수에 대한 gradient 계산시 다른 변수들은 상수로 취급한다.

a = torch.tensor([1.0], requires_grad=True)
b = torch.tensor([2.0], requires_grad=True)

print(a.grad, b.grad)

result = a * b
result.backward()

print(a.grad, b.grad)
```

## torch.no_grad() 
- no_grad()를 with 구문에서 연산을 할 경우 with block 내에서 requires_grad=True로 설정된 변수를 이용한 계산식이 있더라도 grad_fn을 생성하지 않는다. 즉 gradient를 계산하지 않는다.
- 딥러닝 모델을 평가 하거나 추론을할 때는 gradient를 계산할 필요가 없기 때문에 no_grad 구문을 사용한다.


```python
x = torch.tensor([1.0], requires_grad=True)
x.requires_grad = False
y = x * 2
print("결과:", y.data)
print("grad_fn:", y.grad_fn)
```

```python
x = torch.tensor([1.0], requires_grad=True)

with torch.no_grad():
    y = x * 2
    print(y.data, y.grad_fn)
```

## gradient 값 초기화
- backward() 에서 계산되어 저장된 gradient값을 초기화한다.
- 반복 학습을 할 경우 gradient가 누적되는 문제가 발생한다. 그래서 한번 학습이 끝나면 다음 학습 전에 초기화하는 작업을 해야 한다.

```python
x = torch.tensor([1.0], requires_grad=True)
print(x.grad)

y = x ** 2 # grad_fn: 2 * x
y.backward() # grad 계산 : 2 * [1.0]
print(x.grad)
```

```python
# x를 이용한 새로운 계산
x.grad = None
# 새로운 연산을 하기 전에 grad값을 초기화 해야한다. 
# 안하면 이전 grad 값이 남아 있어서 계속 누적 된다.
z = x ** 3 # grad_fn: 3 * x ** 2
z.backward() # x.grad : 3 * 1 ** 2 = 3.
print(x.grad)
```

```python
# 새로운 계산
x.grad = None
r = x ** 2
r.backward()
print(x.grad)
```
